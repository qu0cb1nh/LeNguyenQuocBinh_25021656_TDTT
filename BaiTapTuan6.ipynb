```python
# W6A1
a = list(map(int,input().split()))
m = {}
i = 0
while i < len(a):
    x = a[i]
    if x in m:
        a.pop(i)
    else:
        m[x] = 69
        i += 1
print(a)
```

```python
# W6A3
a = tuple(map(int, input().split()))
k = int(input())
n = len(a)
k %= n
for i in range(k, n):
    print(a[i], end = ' ')
for i in range(0, k):
    print(a[i], end = ' ')
```

```python
# W6A4
inp = input().split()
m = {}
for p in inp:
    key = p.split(":")[0]
    value = p.split(":")[1]
    if key in m:
        m[key].append(value)
    else:
        m[key] = [value]
print(m)
```

```python
# W6A5
a = list(map(int, input().split()))
m = {'positives' : 0, 'negatives': 0, 'zeros' : 0}
for x in a:
    if x > 0:
        m['positives'] += 1
    elif x < 0:
        m['negatives'] += 1
    else:
        m['zeros'] += 1
print(m)
```

```python
# W6A10
a = list(map(int, input().split()))
n = len(a)
k = int(input())
ans = []
for i in range(0, n-1):
    for j in range(i+1, n):
        if a[i] + a[j] == k:
            ans.append((a[i], a[j]))
ans.sort()
print(ans)
```

```python
# W6A12
a = list(map(int, input().split()))
m = {}
mx = 0
mn = -42
for x in a:
    if x in m:
        m[x] += 1
    else:
        m[x] = 1
    cnt = m[x]
    if cnt > mx:
        mx = cnt
        mn = x
    elif cnt == mx:
        mn = min(mn, x)
print(mn, mx)
```
