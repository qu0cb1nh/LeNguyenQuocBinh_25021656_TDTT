```python
# W4A6
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
        if i * i > n:
            break
    return True
    
a, b = map(int, input().split())
ans = 0
for i in range(a, b+1):
    if is_prime(i):
        ans += i
print(ans)
```

```python
# W4A7
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
        if i * i > n:
            break
    return True
    
n = int(input())
ans = 0
for i in range(2, n+1):
    if is_prime(i) and n % i == 0:
        ans = i
print(ans)
```

```python
# W4A8
def reverse(n):
    x = 0
    while n > 0:
        x = x * 10 + (n % 10)
        n //= 10
    return x

def is_palindrome(n):
    s = str(n)
    length = len(s)
    for i in range(length // 2):
        if s[i] != s[length-i-1]:
            return False
    return True

n = int(input())
cnt = 0
while True:
    n = n + reverse(n)
    cnt += 1
    if is_palindrome(n):
        break
print(f'{cnt} buoc, so cuoi cung la {n}')
```

```python
# W4A9
from math import sqrt, isclose
def is_square(x):
    return isclose(sqrt(x), int(sqrt(x)))

def distinct(x):
    check = [False] * 10
    while x > 0:
        d = x % 10
        if check[d] == False:
            check[d] = True
        else:
            return False
        x //= 10
    return True

n = int(input())
for i in range(1, n+1):
    if is_square(i) and distinct(i):
        print(i, end = " ")
```

```python
# W4A11
n = int(input())
cnt = 0
for i in range(2,n+1,2):
    if n % i == 0:
        cnt += 1
        j = n/i
        if j % 2 == 0 and j != i:
            cnt += 1
print(cnt)
```

```python
# W4A13
def tong(n):
    uoc = 0
    for i in range(1,n):
        if n % i == 0:
            uoc += i
    return uoc
x, y = map(int, input().split())
print("true" if tong(x) == y and tong(y) == x else "false")
```

```python
# W4A14
def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a % b)
x, y = map(int, input().split())
print(gcd(x,y))
```
