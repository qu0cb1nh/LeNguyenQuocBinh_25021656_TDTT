```python
# W5A1
def max_of_two(a, b):
    if a > b:
        return a
    return b
x, y = map(int, input().split())
print(max_of_two(x,y))
```

```python
# W5A2
def swap(a, b):
    t = a
    a = b
    b = t
    return (a,b)
x, y = map(int, input().split())
x, y = swap(x,y)
print(x, y)
```

```python
# W5A3
def prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
n = int(input())
print("True" if prime(n) else "False")
```

```python
# W5A4
def hoanhao(n):
    tong = 0
    for i in range(1, n):
        if n % i == 0:
            tong += i
    return tong == n
n = int(input())
print("True" if hoanhao(n) else "False")
```

```python
# W5A5
def find(a, k):
    flag = False
    for i in range(len(a)):
        if a[i] == k:
            print(i+1)
            flag = True
            break
    if flag == False:
        print(-1)

a = list(map(int, input().split()))
k = int(input())
find(a, k)
```

```python
# W5A6
def fact(n):
    prod = 1
    for i in range(2, n+1):
        prod *= i
    return prod
n = int(input())
print(fact(n))
```

```python
# W5A8
def hamming(x, y):
    cnt = 0
    for i in range(0, 65):
        if (x >> i) & 1 != (y >> i) & 1:
            cnt += 1
    return cnt
x, y = map(int, input().split())
print(hamming(x, y))
```

```python
# W5A9
def tongchuso(x):
    tong = 0
    while x > 0:
        tong += x % 10
        x //= 10
    return tong
n = int(input())
print(tongchuso(n))
```

```python
# W5A10
m = {}
m1 = {}
def dangcau(a, b):
    for i in range(len(b)):
        if b[i] in m:
            if m[b[i]] != a[i]:
                return False
        else:
            if a[i] in m1:
                return False
            else:
                m[b[i]] = a[i]
                m[a[i]] = b[i]
    return True
a, b = input().split()
print("True" if dangcau(a,b) else "False")
```
